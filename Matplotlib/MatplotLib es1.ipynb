# Introduction to Matplotlib
Matplotlib: Il Tuo Strumento per Visualizzare i Dati in Python
Matplotlib è una libreria Python estremamente versatile e potente, appositamente progettata per creare visualizzazioni di dati statici, animati e interattivi di alta qualità. È uno strumento essenziale per data scientist, analisti e chiunque voglia comprendere i propri dati attraverso rappresentazioni grafiche.

A cosa serve Matplotlib?

- Visualizzazione esplorativa: Ti permette di esplorare i tuoi dati in modo visivo, cercando pattern, anomalie e relazioni nascoste.
- Comunicazione dei risultati: Puoi creare grafici chiari e concisi per presentare i tuoi risultati a un pubblico non tecnico.
- Pubblicazione: I grafici generati con Matplotlib possono essere facilmente inclusi in report, articoli scientifici e presentazioni.
con %matplotlib inline posso mostrare i grafici direttamente nella finestra di Jupyter

# Figure, Subplot e Plot: Cosa Sono?
Figure: Immagina una figura come una tela vuota su cui andrai a dipingere i tuoi grafici. È il contenitore principale che contiene tutti gli elementi del grafico.
Subplot: Un subplot è come una singola immagine all'interno di una cornice più grande (la figura). Ti permette di dividere la figura in una griglia e di creare più grafici all'interno di ciascuna cella.
Plot: Il plot è il grafico stesso, cioè la rappresentazione visiva dei tuoi dati. Può essere una linea, un punto, una barra, ecc.
Un Esempio Pratico: Confrontare Vendite Mensili di Diversi Prodotti
Supponiamo di avere un dataset che contiene le vendite mensili di tre prodotti diversi. Vogliamo creare un grafico che mostri il trend delle vendite per ciascun prodotto sullo stesso grafico, per facilitare il confronto.

```python

# Simuliamo i dati
np.random.seed(0)
mesi = pd.date_range('2023-01-01', periods=12, freq='ME')
prodotto_A = np.random.randint(100, 200, size=12)
prodotto_B = np.random.randint(150, 250, size=12)
prodotto_C = np.random.randint(200, 300, size=12)

# Creiamo una figura con un subplot
plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)  # 1 riga, 1 colonna, subplot 1
plt.plot(mesi, prodotto_A, label='Prodotto A')
plt.plot(mesi, prodotto_B, label='Prodotto B')
plt.plot(mesi, prodotto_C, label='Prodotto C')
plt.xlabel('Mesi')
plt.ylabel('Vendite')
plt.title('Confronto Vendite Mensili')
plt.legend()
plt.grid(True)
plt.show()
```

# Questo aiuta a capire perchè subplot (1,1,1)
Adesso creo due subplot, uno affiancato all'altro
(come creare una stanza grande con un solo mobile)

```python
# Dati casuali per due grafici
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creiamo una figura con due subplot
plt.figure(figsize=(12, 6))

# Primo subplot
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Grafico del seno')

# Secondo subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Grafico del coseno')

plt.show()
```

# Un plot da un dizionario

```python
nut_butter_price = {
    "Almond butter": 10, "Peanut butter": 8, "Cashes butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()),list( nut_butter_price.values()))
```
