```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
hd = pd.read_csv("heart-disease.csv")
hd["age"].plot.hist(bins=20)
```

```python
# Crea un DataFrame con dati casuali sulle vendite
np.random.seed(42)
data = {'Prodotto A': np.random.randint(10, 100, 12),
        'Prodotto B': np.random.randint(50, 150, 12)}
df = pd.DataFrame(data)

# Crea un istogramma senza subplots
plt.hist(df)
plt.xlabel('Vendite mensili')
plt.ylabel('Frequenza')
plt.title('Distribuzione delle vendite')
plt.show()
```

```python
over_50 = hd[hd["age"]> 50]
over_50.plot(kind='scatter', x = 'age', y = 'chol', c ='target')
```

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y = 'chol',c='target',ax=ax)
ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
# Cos posso castomizzare il plot
ax.set(title='Heart deasese and Cholesterl levels', xlabel='Age', ylabel='Colesterolo')

# Cosi posso aggiungere una leggenda
ax.legend(*scatter.legend_elements(),title='Target')

# Come aggiungere una linea per avere la media del colesterolo
ax.axhline(over_50['chol'].mean(), linestyle='--')

```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1, figsize=(10,10))

scatter = ax0.scatter(x=over_50['age'], y = over_50['chol'], c=over_50['target'])
ax0.set(title='Heart deasese e livello di colesterelo', xlabel='Et', ylabel='Colesterolo')
ax0.legend(*scatter.legend_elements(), title='target')
ax0.axhline(y=over_50['chol'].mean(),linestyle='--')

scatter = ax1.scatter(x=over_50['age'], y = over_50['thalach'], c=over_50['target'])
ax1.set(title='Heart deasese and max heart rate', xlabel='Et', ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--')

#Adesso aggiungo un titolo alla figura
fig.suptitle("Heart Disease analysis", fontsize=16, fontweight='bold')
```
