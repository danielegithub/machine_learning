## Manipulating data (missing data)

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
car_sales = pd.read_csv("car-sales.csv");
```

```python
car_sales
```

```python
car_sales["Make"].str.lower()
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# fillna - riempe i valori nulli con dati
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
car_sales_missing = car_sales_missing.dropna()
```

```python
car_sales_missing
```

```python
seats_column = pd.Series([5,5,5,5,5])
# Aggiungo una nuova colonna
car_sales["Seats"]=seats_column
car_sales
```

```python
car_sales["Seats"] = car_sales["Seats"].fillna(5)
```

```python
car_sales
```

```python
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# Si pu√≤ anche creare una colonna da un singolo valore
car_sales["Number of wheels"] = 4
car_sales["Passed road saftey"] = True
```

```python
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used", axis=1)
```

```python
car_sales
```

```python
# Come mescolare i dati
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
car_sales_shuffled.sample(frac=0.01)
```

```python

car_sales = car_sales.reset_index(drop=True)
car_sales
```

```python
# Per applicare una funzione
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/16)
```
